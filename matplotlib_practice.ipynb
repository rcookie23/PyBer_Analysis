```python
# Import dependencies
import matplotlib.pyplot as plt 
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, label='Boston', c='g', linewidth=2, marker = "d" )
ax.grid()
ax.legend()
ax.set_title("Pyber Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_ylim(0, 45)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker = "*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Create a vertical bar chart using the pyplot
# Create the plot
plt.bar(x_axis, y_axis, color="green", label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Create a horizontal bar chart using pyplot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.legend
plt.title("Pyber Fare by Month")
plt.xlabel("Fares ($)")
plt.ylabel("Date")
```

```python
# Create a vertical bar chart using the object oriented method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
plt.scatter(y_axis, x_axis, c="red", label = 'Chicago')
plt.legend()
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.xlim(0, 45)
plt.gca().invert_yaxis()
plt.title("Pyber Fare by Month")
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], c="skyblue", alpha = 0.8, edgecolors="k", lw=2, label = "Boston")
ax.legend()
ax.invert_yaxis()
ax.set_title("Pyber Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots()
fig = plt.gcf()
fig.set_size_inches(8,8)
plt.subplots(figsize=(8, 8))
ax.pie(y_axis, labels=x_axis, autopct='%.2f%%', colors=colors, explode=explode_values, shadow=True, counterclock = False, startangle = 90)
plt.show
```
